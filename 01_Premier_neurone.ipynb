```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

# 1. Dataset

```python
X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=0)
# X[:, 1] = X[:, 1] * 10

y = y.reshape((y.shape[0], 1))

print('dimensions de X:', X.shape)
print('dimensions de y:', y.shape)

plt.scatter(X[:,0], X[:, 1], c=y, cmap='summer')
```

# 2. Fonctions du modele

```python
def initialisation(X):
    W = np.random.randn(X.shape[1], 1)
    b = np.random.randn(1)
    return (W, b)
```

```python
def model(X, W, b):
    Z = X.dot(W) + b
    A = 1 / (1 + np.exp(-Z))
    return A
```

```python
def log_loss(A, y):
    epsilon = 1e-15
    return 1 / len (y) * np.sum(-y * np.log(A + epsilon) - (1 - y) * np.log(1 - A + epsilon))
```

```python
def gradients(A, X, y):
    dW = 1 / len(y) * np.dot(X.T, A - y)
    db = 1 / len(y) * np.sum(A - y)
    return (dW, db)
```

```python
def update(dW, db, W, b, learning_rate):
    W = W - learning_rate * dW
    b = b - learning_rate * db
    return (W, b)
```

```python
def predict(X, W, b):
    A = model(X, W, b)
    #print(A)
    return A >= 0.5
```

```python
from sklearn.metrics import accuracy_score
```

```python
def artificial_neuron(X, y, learning_rate=0.1, n_iter=100) :
    
    W, b = initialisation(X)
    
    Loss = [] 
    acc = []
    
    # Training
    for i in range (n_iter):
        # Activation
        A = model(X, W, b)
        
        # Caclul du cout
        Loss.append (log_loss (y, A))
        
        #Calcul de la précision
        y_pred = predict(X, W, b)
        acc.append(accuracy_score(y, y_pred))
        
        #Mise à jour
        dW, db = gradients (A, X, y)
        W, b = update(dW, db, W, b, learning_rate = learning_rate)
        
    plt.figure(figsize=(12,4)) 
    plt.subplot(1,2,1)
    plt.plot(Loss)
    plt.subplot(1,2,2)
    plt.plot(acc)
    plt.show()
    return (W,b)
```

```python
W, b = artificial_neuron(X, y)
```

```python
def artificial_neuron_2(X, y, learning_rate=0.1, n_iter=100) :
    
    W, b = initialisation(X)
    W[0], W[1] = -7.5, -7.5
    
    nb = 10
    j = 0 
    history = np.zeros((n_iter // nb, 5))
    Loss = [] 
    
    # Training
    for i in range (n_iter):
        A = model(X, W, b)
        Loss.append(log_loss (y, A))
        dW, db = gradients(A, X, y)
        W, b = update(dW, db, W, b, learning_rate = learning_rate)
        
        if (i % nb == 0):
            history[j, 0] = W[0]
            history[j, 1] = W[1]
            history[j, 2] = b
            history[j, 3] = i
            history[j, 4] = log_loss (y, A)
            j +=1
        
    plt.plot(Loss)
    plt.show()
    return history, b
```

```python
history, b = artificial_neuron_2(X, y)
```
